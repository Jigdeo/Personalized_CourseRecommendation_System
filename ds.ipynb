```python
import random
import pandas as pd

# Set the number of courses to generate
num_courses = 5000
courses=pd.read_csv('courses_list.csv')
teachers=pd.read_csv('Indian_Names.csv',index_col = [0])
#teachers=teachers1.sample(50)
x=len(courses)
y=len(teachers)
courses.head()
```

```python

# Generate the course data
fcourses = []
for i in range(num_courses):
    a=random.randint(1,x)-1
    b=random.randint(1,y)-1
    sno=i + 1
    course_id = courses["Course Code"][a]
    course_name =  courses["Course Title"][a]
    teacher_name =   teachers["Name"][b]
    teacher_name = "Dr. "+teacher_name
    teacher_rating = random.randint(1, 5)
    teacher_friendliness = random.randint(1, 5)
    num_credits = random.randint(2, 5)
    course_difficulty = random.randint(1, 5)%num_credits+(random.randint(1,5)%random.randint(1,3))
    teacher_teaching_skills = random.randint(1, 5)
    
    fcourses.append([sno,course_id, course_name, teacher_name, teacher_rating, teacher_friendliness, course_difficulty, teacher_teaching_skills, num_credits])

#
# Set the number of courses to generate
#num_students = 100


# Generate the course data

students = []
for i in range(5000):
    sno=i + 1
    student_id = i + 1
    credits = random.randint(1,5)
    cgpa = random.randint(0,10)
    course_diff=credits-1.5*(random.randint(0,2))+1.5*(random.randint(0,2))

    if cgpa >= 2*course_diff-1.5*(random.randint(0,2)):
        pf=1
    else:
        pf=0   
    students.append([sno,student_id,credits,cgpa,course_diff,pf])


# Convert the data to a Pandas DataFrame
df1= pd.DataFrame(fcourses, columns=['Serial No.','Course ID', 'Course Name', 'Teacher Name', 'Teacher Rating', 'Teacher Friendliness', 'Course Difficulty', 'Teacher Teaching Skills', 'Number of Credits'])
df2= pd.DataFrame(students, columns=['Serial No.','Student ID','Credits', 'CGPA', 'Course Difficulty', 'PASS/FAIL'])

# Save the data to a CSV file
df1.to_csv('courses.csv', index=False)
df2.to_csv('students.csv', index=False)

# Display 5 random samples from the dataset
df1.sample(5)
df2.sample(5)

```
